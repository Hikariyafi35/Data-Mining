```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = pd.read_csv("Factory_Salary.csv")
dataset.head()
```

```python
dataset = dataset.drop(['Date'],axis =1)
dataset
```

```python
dataset.columns
```

```python
dataset.Profession.unique()
```

```python
dataset.info()
```

```python
dataset['Equipment'] = dataset['Equipment'].str.strip()

```

```python
#gaji di pangkat refractory
basket_Refractory = (dataset[dataset['Profession']=="Refractory"].groupby(['Rank','Equipment'])['Insalubrity'].sum().unstack().reset_index().fillna(0).set_index('Rank'))
#gaji di pangkat roler
basket_Roller = (dataset[dataset['Profession']=="Roller"].groupby(['Rank','Equipment'])['Insalubrity'].sum().unstack().reset_index().fillna(0).set_index('Rank'))
#gaji di pangkat cleaner
basket_Cleaner = (dataset[dataset['Profession']=="Cleaner"].groupby(['Rank','Equipment'])['Insalubrity'].sum().unstack().reset_index().fillna(0).set_index('Rank'))
#gaji di pangkat Operator
basket_Operator = (dataset[dataset['Profession']=="Operator"].groupby(['Rank','Equipment'])['Insalubrity'].sum().unstack().reset_index().fillna(0).set_index('Rank'))

Operator
```

```python
def hot_encode(x):
  if(x<= 0 ):
    return 0
    if(x>=1):
      return 1
```

```python
basket_encoded = basket_Refractory.applymap(hot_encode)
basket_Refractory = basket_encoded

basket_encoded = basket_Roller.applymap(hot_encode)
basket_Roller = basket_encoded

basket_encoded = basket_Cleaner.applymap(hot_encode)
basket_Cleaner = basket_encoded

basket_encoded = basket_Operator.applymap(hot_encode)
basket_Operator = basket_encoded

basket_Operator
```

```python
frq_items = apriori(basket_Refractory, min_support = 0.05, use_colnames = True)

rules = association_rules(frq_items, metric = "lift", min_threshold= 1)
rules = rules.sort_values(['confidence','lift'], ascending = [false,false])
print(rules.head())
```

```python
frq_items = apriori(basket_Roller, min_support = 0.05, use_colnames = True)

rules = association_rules(frq_items, metric = "lift", min_threshold= 1)
rules = rules.sort_values(['confidence','lift'], ascending = [false,false])
print(rules.head())
```

```python
frq_items = apriori(basket_Cleaner, min_support = 0.05, use_colnames = True)

rules = association_rules(frq_items, metric = "lift", min_threshold= 1)
rules = rules.sort_values(['confidence','lift'], ascending = [false,false])
print(rules.head())
```

```python
frq_items = apriori(basket_Operator, min_support = 0.05, use_colnames = True)

rules = association_rules(frq_items, metric = "lift", min_threshold= 1)
rules = rules.sort_values(['confidence','lift'], ascending = [false,false])
print(rules.head())
```
